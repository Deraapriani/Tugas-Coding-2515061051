Vector

```python
import numpy as np
```

```python
# Membuat array NumPy (Vektor) Argumen tipe data bersifat opsional, contoh: dtype=np.float16
x = np.array([20, 4, 6])
x
```

```python
# Hitung panjang vector x
print(len(x))
```

```python
# Tampilkan ukuran vector x
print("Ukuran vector x:", x.shape)
```

```python
# Tampilkan tipe data dari vector x
print("Tipe data vector x:", type(x))
```

```python
# Akses elemen pertama dari vector x
print("Elemen pertama:", x[0])
```

```python
# Tampilkan tipe data elemen pertama dari vector x
print("Tipe data elemen pertama:", type(x[0]))
```

Vector Transposition

```python
# Buat vector baris x (ukuran 1x3) agar bisa ditranspose menjadi 3x1
x_t = x.T
x_t
```

```python
# Tampilkan ukuran dari vector x_t
x_t.shape
```

```python
# Buat vector y (ukuran 1x3)
y = np.array([[13, 10, 25]])
y
```

```python
# Tampilkan ukuran dari vector x_t
y.shape
```

```python
# kita bisa melakukan transpose pada matriks yang salah satu dimensinya bernilai 1, yang secara matematis setara:
y_t = y.T
y_t
```

```python
y_t.shape # # ini adalah vektor kolom karena memiliki 3 baris dan 1 kolom
```

```python
# Vektor kolom dapat di-transpose kembali menjadi vektor baris aslinya:
y_t.T
```

```python
y_t.T.shape
```

vector dengan tensorflow

```python
import tensorflow as tf
```

```python
# # lakukan pada tahapan ini dengan tensorflow
#Buat variabel(1x3)
x_tf = tf.Variable([[20, 4, 6]])
x_tf
```

```python
y_tf2 = tf.Variable([[20, 4, 6]])
y_tf2
```

```python
x_tf.shape
```

```python
y_tf = tf.Variable([[20, 4, 6]])
print(f"variabel tensor y adalah :", y_tf)
```

```python
x_tf + y_tf
```

```python
tf_sum = tf.add(x_tf, y_tf)
tf_sum
```

```python
tf_sum.numpy() # perhatikan bahwa operasi NumPy secara otomatis mengubah tensor menjadi array NumPy, dan sebaliknya
```
