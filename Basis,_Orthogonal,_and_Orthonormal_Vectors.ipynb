Orthogonal Vectors

```python
import numpy as np
```

```python
i = np.array([1, 0])
i
```

```python
j = np.array([0, 1])
j
```

```python
np.dot(i, j) # detail on the dot operation coming up...
```

```python
# Bukti Orthogonal: Hitung Dot Product antara i dan j
# Rumus: (1*0) + (0*1) = 0
dot_product = np.dot(i, j)

print("Dot Product i dan j:", dot_product)
```

```python
if dot_product == 0:
    print("Kesimpulan: Vektor i dan j adalah Orthogonal.")
else:
    print("Kesimpulan: Vektor i dan j tidak Orthogonal.")
```

Orthogonal Vectors

```python
import numpy as np
```

```python
# Membuat vektor v
v = np.array([1, 2])
v
```

```python
# Membuat vektor w
w = np.array([2, -1])
w
```

```python
# Bukti Orthogonal: Hitung Dot Product antara v dan w
# Rumus: (1*2) + (-2*1) = 0
dot_product = np.dot(v, w)

print("Dot Product v dan w:", dot_product)
```

```python
if dot_product == 0:
    print("Kesimpulan: Vektor i dan j adalah Orthogonal.")
else:
    print("Kesimpulan: Vektor i dan j tidak Orthogonal.")
```
