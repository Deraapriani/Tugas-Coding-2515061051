```python
# Definisi kolom vektor (Matrix 4x1)
vektor = [
    [25],
    [2],
    [-3],
    [-23]
]
```

```python
# Proses Transpose: Mengubah baris menjadi kolom
# Kita mengambil elemen pertama dari setiap sub-list
transpose = [baris[0] for baris in vektor]

print("Vektor Asal (Kolom):")
for b in vektor:
    print(b)

print("\nHasil Transpose (Baris):")
print(transpose)
```
